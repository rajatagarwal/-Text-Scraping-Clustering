## Loading Corpus

```python
# Text corpus path
file_path = "data_dump/text_data.txt"
# Open file and store into text_data
text_data = open(file_path,"r")
# reading text_data line by line and storing in "documents". "documents" is a list, 
# where each element represents an individual article
documents = text_data.readlines()
# close file
text_data.close()
# displayes total number of articles in the "documents" list
print("Read %d raw text documents" % len(documents))
```

## Text Preprocessing and Document-Term Matrix Representation

```python
# define the function
def lemma_tokenizer(text):
    # use the standard scikit-learn tokenizer first
    standard_tokenizer = CountVectorizer().build_tokenizer()
    tokens = standard_tokenizer(text)
    # then use NLTK to perform lemmatisation on each token
    lemmatizer = nltk.stem.WordNetLemmatizer()
    lemma_tokens = []
    for token in tokens:
        lemma_tokens.append( lemmatizer.lemmatize(token) )
    return lemma_tokens
```

```python
# It will remove stop words, get words with more than 3 frequency, set ngram to unigram and bigrams, 
# and lemmatize using WordNet Lemmatizer. Once done, converting it to the TFIDF document frequency
vectorizer = TfidfVectorizer(stop_words="english",min_df = 3,tokenizer=lemma_tokenizer, ngram_range=(1,2))
# Fit the vectorizer to the documents list.
X = vectorizer.fit_transform(documents)
# X is as scipy.sparse.csr.csr_matrix
# We can check size of X
print(X.shape)
# terms is a list. terms = all the keywords (features)
terms = vectorizer.get_feature_names()
# vocab is a dictionary. It maps terms to feature indices.
vocab = vectorizer.vocabulary_
print("Vocabulary has %d distinct terms" % len(terms))
```

```python
# Creating a pandas dataframe based on terms, which is used later in clustering
vocab_dataframe = pd.DataFrame({'words': terms}, index = terms)
```

# TASK 2

Following tasks are completed:
* Find most characterstic terms in the corpus
* Find most characterstic phrases in the corpus

```python
# Get top n tfidf values in row and return them with their corresponding feature names.
def top_tfidf_feats(row, features, top_n):
    rank = list(range(1,top_n+1))
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.insert(0, "rank", rank)
    df.columns = ['rank', 'term', 'tfidf score']
    return df
```

```python
# Return the top n features that on average are most 
# important amongst documents in rows indentified by indices in grp_ids.
def top_mean_feats(X, features, top_n, grp_ids=None, min_tfidf=0.1):
    if grp_ids:
        D = X[grp_ids].toarray()
    else:
        D = X.toarray()
    D[D < min_tfidf] = 0
    tfidf_means = np.mean(D, axis=0)
    return top_tfidf_feats(tfidf_means, features, top_n)
```

```python
print("Most characterstics terms in the Corpus.")
# Display top 20 most characterstics terms with their tfidf scores, in the Corpus.
display(HTML(top_mean_feats(X_t2_single, terms_t2_single, 20).to_html(index = False)))
```

## Algorithm 1: K-Means Algorithm

```python
# Number of clusters we want. This we choose based on experiments. We want to find all clusters of similar items.
num_of_clusters = 4
# Kmeans model
km_model = KMeans(n_clusters = num_of_clusters)
# Fitting Kmeans model to document-term matrix
km_model.fit(X)
# listing cluster labels
clusters = km_model.labels_.tolist()
print(clusters)
print("Created 4 clusters based on experiment.")
```

```python
# Getting a doc_dict which have each document its corresponding cluster and its document number (generated in step 1)
doc_dict = {'doc_text': documents, 'cluster': clusters, 'doc_number': doc_number, 'title': titles}
# Creating the data frame using above dictionary and indexing based on cluster number
pd_kmeans = pd.DataFrame(doc_dict, index = [clusters], columns = ["cluster", "doc_number", "title"])
```

```python
# Number of documents per cluster
print("Number of documents per cluster:")
pd_kmeans['cluster'].value_counts()
```

```python
# Top terms per clusters
print("\nTop terms per K-Means cluster:\n")
#sort cluster centers by proximity to centroid
order_centroids = km_model.cluster_centers_.argsort()[:, ::-1]
# For all cluster find terms closest to the centroid
for i in range(num_of_clusters):
    print("Cluster %d words:" % i, end='\n')
    # Display top-n terms closest to the centroid
    for ind in order_centroids[i, :6]: #replace 11 with n words per cluster
        print(' %s' % (vocab_dataframe.ix[terms[ind].split(' ')].values.tolist()[0][0].encode('utf-8', 'ignore')).decode('utf-8'))
    print()
```

## Algorithm 2: Agglomerative Clustering Algorithm

```python
# Calculating the cosine similarity based on the TFIDF Score
distance = 1 - cosine_similarity(X)
```

```python
#define the linkage_matrix using ward clustering pre-computed distances
lm = ward(distance) 
# setting dendrogram size
figure, dendogram_ax = plt.subplots(figsize=(15, 200)) 
print("It created 4 clusters of documents.")
# Dendrogram configurations settings
dendogram_ax = dendrogram(lm, orientation="right", labels=titles)
plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',       # both major and minor ticks are affected
    bottom='off',       # ticks along the bottom edge are off
    top='off',          # ticks along the top edge are off
    labelbottom='off')
#show plot with tight layout
plt.tight_layout() 
```

## Summarise the contents of the clusters

* Summarising titles for each clusters (Each cluster is based on documents)
* Finding most popular titles based on score calcuated on tokens score
* Each token score is based on its occurrences in the clustered titles

```python
# remove non ascii values from string
def removeNonAscii(s): return "".join(i for i in s if ord(i)<128)
```

```python
def summarise_titles(titles):
    non_ascii_title = []
    # remove non ascii values
    for title in titles:
        non_ascii_title.append(removeNonAscii(title)) 
    
    # Defining nltk's wordpunct tokenizer
    tokenizer = WordPunctTokenizer()
    # 'sentences' will a list of the tokenized sentences
    tokenise_titles = []
    for title in non_ascii_title:
        # Tokenize each sentence
        tokens = tokenizer.tokenize(title.lower())
        # Add the tokenized sentence to our list of sentences
        tokenise_titles.append(tokens)

    # To count all the words (tokens) in the document, we must iterate:
    # 1) over each sentence, and then
    # 2) over each word (token) in the sentence
    # And we'll store the counts in a 'Counter' object.
    token_counts = Counter()
    for title in tokenise_titles:
        for word in title:
            token_counts[word] += 1

    # remove stop words from tokens
    stopwords = nltk.corpus.stopwords.words('english')
    for token in token_counts:
        # Set stopword counts to zero
        if token in stopwords:
            token_counts[token] = 0
        # Set non-alphanumeric tokens (e.g. puncutation) counts to zero
        elif not token[0].isalpha():
            token_counts[token] = 0
    
    # score sentences
    scores = []
    for sentence in tokenise_titles:
        score = 0
        for token in sentence:
            score += token_counts[token]
        scores.append(score)

    # we will bundle score together
    score_dict = {'score': scores, 'title': non_ascii_title}
    # Creating the data frame using above dictionary and indexing based on cluster number
    score_df = pd.DataFrame(score_dict, index = [scores], columns = ["score", "title"])
    score_df.sort_index(inplace=True, ascending= False)
    
    # Get dataframe values into list
    sorted_titles = score_df.ix[:, 1].tolist()
    
    # Remove duplicate titles
    top_titles = list(set(sorted_titles))
    
    for item in top_titles[:5]:
        print(item)
```

```python
# Process documents present in each clusters
for i in range(num_of_clusters):
    # Summarizing cluster based on top titles. Top titles are calculated based on most frequent terms
    print("Summary of cluster #", i, "based on top 5 titles of News\n")
    titles = []
    for title in pd_kmeans.ix[i]['title'].values.tolist():
        titles.append(title)
    # Summarising titles. 
    summarise_titles(titles)
    print()
```
